```python
temp_price = price.groupby('ticker')    
tickers = temp_price.groups
```

```python
tickers.keys()
```

```python
returns = pd.DataFrame(columns=['ticker', 'returns'])

for i in tickers.keys():
    start = temp_price.get_group(i).iloc[[0]]
    end = temp_price.get_group(i).iloc[[-1]]
    start = start['close'].values
    end = end['close'].values
    r = float((start-end)*100/start)
    new_row = pd.DataFrame({'ticker':[i], 'returns':[r]})
    returns = pd.concat([new_row,returns.loc[:]])
```
